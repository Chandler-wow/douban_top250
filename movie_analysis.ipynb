```python
import pandas as pd
import matplotlib.pyplot  as plt
from matplotlib.ticker import MultipleLocator
```

```python
movies = pd.read_csv("./assets/movies.csv")
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure()
fig ,ax1 = plt.subplots(figsize=(100,20))
ax1 = plt.gca()
ax1.xaxis.set_major_locator(MultipleLocator(1))
ax1.yaxis.set_major_locator(MultipleLocator(1))
ax1.set_xlabel("排名")
ax1.set_ylabel("上映日期")  
ax1.plot(movies["排名"],movies["上映日期"].apply(lambda x : int(x)),linestyle = "dashed",color = "red",marker = "o",label = "上映日期")
ax1.tick_params(axis="y",labelcolor = "tab:red")
ax2 = ax1.twinx()
ax2.set_ylabel("评分")
ax2.plot(movies["排名"],movies["评分"],color = "blue",marker = "*",label = "评分")
ax2.tick_params(axis="y",labelcolor = "tab:blue")
plt.xlim(1,250)
plt.title("豆瓣Top250 上映日期")
plt.show()
```

```python
df = movies["影片类型"].map(lambda x: x.strip().split(" ")).explode("影片类型").to_frame()
df["影片类型"].value_counts().plot.pie(autopct = "%.2f",fontsize = 10,figsize= (17,17))
plt.title("豆瓣Top250影片类型分布")
plt.show()
```
