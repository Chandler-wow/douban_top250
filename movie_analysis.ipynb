```python
import pandas as pd
import matplotlib.pyplot  as plt
from matplotlib.ticker import MultipleLocator
```

```python
movies = pd.read_csv("./assets/movies.csv")
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure()
fig ,ax1 = plt.subplots(figsize=(100,20))
ax1 = plt.gca()
ax1.xaxis.set_major_locator(MultipleLocator(1))
ax1.yaxis.set_major_locator(MultipleLocator(1))
ax1.set_xlabel("")
ax1.set_ylabel("")  
ax1.plot(movies[""],movies[""].apply(lambda x : int(x)),linestyle = "dashed",color = "red",marker = "o",label = "")
ax1.tick_params(axis="y",labelcolor = "tab:red")
ax2 = ax1.twinx()
ax2.set_ylabel("")
ax2.plot(movies[""],movies[""],color = "blue",marker = "*",label = "")
ax2.tick_params(axis="y",labelcolor = "tab:blue")
plt.xlim(1,250)
plt.title("Top250 ")
plt.show()
```

```python
df = movies[""].map(lambda x: x.strip().split(" ")).explode("").to_frame()
df[""].value_counts().plot.pie(autopct = "%.2f",fontsize = 10,figsize= (17,17))
plt.title("Top250")
plt.show()
```
